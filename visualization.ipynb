## import libs

```python
import seaborn as sns
import os 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# import clean data spreadsheet

cleanData = pd.read_excel("cleanDataWithProfit.xlsx")
cleanData.info()
```

```python
cleanData.isnull().sum().sum()
```

```python
sns.set()
```

#### Note: Only some of these graphs are used in the final report, however, we decided to keep the others here as well in order to display the thought process.

### *What is the relationship between  gross and cost for movies? Should you invest more to earn more?*  

```python
plt.figure(figsize = (20,20)) #set figure size

#create a line plot
sns.lmplot(data=cleanData, x="Gross (in USD)", y="Cost (in USD)", height = 9)

#set title
plt.title('The Relationship between Gross and Cost', fontsize = 18, fontweight = "bold")

#set labels
plt.xlabel('Gross (in USD)', fontsize = 15)
plt.ylabel('Cost (in USD)', fontsize = 15)

plt.show()
```

### *How does being available on Netflix affect the gross?*

```python
plt.figure(figsize = (15,15)) # set figure size

# create a categorical box plot
sns.catplot(data=cleanData, x="Gross (in USD)", y="Available on Netflix?", kind="box")
#set title
plt.title('The Gross of Movies by Availability on Netflix',fontsize = 20, fontweight = "bold")

#set labels
plt.xlabel('Gross (in USD)', fontsize = 15)
plt.ylabel('Is the movie on Netflix?', fontsize = 15)

plt.show()
```

## *How do the number of the viewers change for different platforms over the years?*

```python
# set a grid 
g = sns.FacetGrid(cleanData, col = "Available on Netflix?", height = 6.5, aspect = 1.0) 

# place a line plot in the grid
g.map(sns.lineplot, "Year", "Number of Viewers (in millions)")

g.set_ylabels(label = "Number of Viewers")

#set spacing for the title
g.fig.subplots_adjust(top=0.9)

#set the title
g.fig.suptitle('Change in viewers on different platforms over the years', fontsize = 15, fontweight = 'bold')
```

### *Do ratings change with different platforms?*

```python
plt.figure(figsize = (15,10)) #set figure size

# set a density plot
sns.kdeplot(data = cleanData , x="Rating (1-10)", hue="Available on Netflix?")

#set title and labels
plt.title('Ratings of the Movies Released in Netflix Compared to Those Released on Other Platforms ',fontsize = 18, fontweight = "bold")
plt.xlabel('Rating (1-10)', fontsize = 16)
plt.ylabel('Density', fontsize = 18)
```

```python
# set figure size
plt.figure(figsize = (15,10))

# create a line plot
sns.lineplot(data=cleanData, x="Year", y="Cost (in USD)")
#set title and labels
plt.title("Change of Cost in Movie Business over the Years", fontsize = 20, fontweight = "bold")
plt.xlabel('Year', fontsize = 20)
plt.ylabel('Cost (in USD)', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (15,10)) #set figure size

# create a line plot
sns.lineplot(data=cleanData, x="Year", y="Number of Viewers", hue = 'Venue Outside of US?')

# set title and labels
plt.title("Does filming outside of US affect views?", fontsize = 20, fontweight = "bold")
plt.xlabel('Year', fontsize = 20)
plt.ylabel('Number of Viewers', fontsize = 20)

plt.show()
```
